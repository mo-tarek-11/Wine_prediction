```python
#Import libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.svm import SVC
from xgboost import XGBClassifier   #Use it to achieve high accuracy on predictions.
from sklearn.linear_model import LogisticRegression

#To turn off warning messages.
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import dataset.
data = pd.read_csv('winequality-red.csv')
data.head()
```

```python
#Create best quality column to use it in prediction for high accuracy
data['top_qu'] = [1 if x > 5 else 0 for x in data.quality]
X = data.drop(['quality', 'top_qu'], axis=1)
y = data['top_qu']
```

```python
X
```

```python
y
```

```python
#Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Normalise
norm = MinMaxScaler()
X_train = norm.fit_transform(X_train)
X_test = norm.transform(X_test)
```

```python
#The y_train must be encoded in a newer update XGBoost model before training it
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
#Training models.
models = [LogisticRegression(random_state = 0),XGBClassifier() ,SVC(kernel='rbf')]
for i in range(3):
    models[i].fit(X_train, y_train)
```

```python
#Predict and show the Accuracy & Confutin matrix
for i in range (3):
    y_pred = models[i].predict(X_test)
    y_pred = le.inverse_transform(y_pred)
    cm = confusion_matrix(y_test, y_pred)
    print("Confution matrix for model " f'{i+1} : \n',cm)
    ac = accuracy_score(y_test, y_pred)
    print("Accuracy score for model " f'{i+1} : ',ac)
```
